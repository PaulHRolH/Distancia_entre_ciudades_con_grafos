```python
import networkx as nx
import matplotlib.pyplot as plt

def construir_grafo(archivo):
    G = nx.Graph()
    with open(archivo,'r') as file:
        next(file)
        for line in file:
            origen, destino, distancia = line.strip().split(',')
            G.add_edge(origen.strip(),destino.strip(), weight = float(distancia.strip()))
    return G
archivo = 'informacion_carreteras.txt'
grafo = construir_grafo(archivo)
print("Nodos del grafo:")
print(grafo.nodes())

print("\nAristas del grafo:")
print(grafo.edges(data=True))



# Dibuja el grafo
plt.figure(figsize=(20, 12))
pos = nx.spring_layout(grafo)  # Define la posición de los nodos
nx.draw(grafo, pos, with_labels=True, node_size=1000, node_color="skyblue", font_size=10)  # Dibuja el grafo
labels = nx.get_edge_attributes(grafo, 'weight')  # Obtiene las etiquetas de distancia
nx.draw_networkx_edge_labels(grafo, pos, edge_labels=labels)  # Dibuja las etiquetas de distancia
plt.title("Grafo de Conexiones entre Ciudades")
plt.show()
```
